# Прогнозирование оттока клиентов - отчет о проведенном исследовании

## Цели и задачи проекта

**Целью проекта** является построение прогнозной модели для телеком-компании, позволяющей предсказывать склонность клиента к оттоку. 

В данном случае решается задача классификации пользователей на два класса "склонный к оттоку" и "не склонный к оттоку".


**Для достижения цели были решены следующие задачи:**
* был проведен описательный анализ имеющихся данных
* проведена подготовка данных для дальнейшего построения модели
* были построены несколько baseline-решений и выбрана наилучшая модель
* проведен подбор оптимальных параметров модели
* оценен экономический эффект от внедрения модели

Полученная модель может быть применена для выделения сегмента пользователей, склонных к оттоку, с которыми затем проводятся целевые коммуникации с целью удержания.

## Методика измерения качества и критерий успех

Для оценки качества модели применялась метрика ROC-AUC. Преимуществом данной метрики является то, что она устойчива к несбалансированным классам. Критерием качества считается значение выше 50%, так как в данном случае мы уже предсказываем отток не случайно.
Качество модели оценивалось с помощью стратегии кросс-валидации (stratified k-fold для сохранения исходного соотношения классов) с 3 фолдами. Ткже качество оценивалось на отложенной выборке.

## Оценка качества модели

При том, что в данных очень много пропусков, а классы сильно не сбалансированны, точность по целевой метрике 70% на тестовых данных довольно высокая.

<img src="learning curves.png">

Кроме того, заметен экономический эффект от внедрения модели. Прибыль от внедрения была рассчитана по следующей формуле:

\begin{eqnarray}Profit =  RPC * p * TP - N * top_pec * RetCosts \end{eqnarray}
    
$RPC$ - средняя выручка на одного клиента

$p$ - вероятность принятия предложения

$TP$ - число верно классифицированных клиентов, склонных к оттоку

$N$ - число клиентов

$top_pec$ - процент клиентов, к которым мы выбираем применить политику удержания

$RetCosts$ - средние затравты на удержание одного клиента


Выявлено, что при заданных параметрах $RPC$ = 100 у.е., $RetCosts$ = 15 у.е. и $p$ = 0.7 оптимальным является выбор 6% из топа склонных к оттоку клиентов, согласно предсказанию модели. Выручка в данном случае составляет 3530 у.е.

## Итог работы

В результате проекта была проделана большая работа: были определены оптимальные способы предобработки данных, а также построена модель, которая с точностью 70% предсказывает отток клиентов телком-компании. Кроме того, был рассчитан экономический эффект от внедрения модели, который показал, что с ее помощью компания может получить значительную экономию. 

Для проверки работспособности модели перед ее внедрением также неплохо провести A/B-тестирование, в котором к части пользователей, склонных к оттоку согласно моедли ,будут примеены меры по удержанию, а к другой нет. Это позволит проверить как точность модели, так и экономический эффект от данного решения
