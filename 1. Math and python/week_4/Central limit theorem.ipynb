```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

# Gamma distribution

The probability density function for gamma is $ f(x,a) = \frac{x^{a-1}  e^{-x}}{Г(a)} $, 

for $x \geq 0, a \geq 0$. Here $Г(a)$ refers to the gamma function.

Gamma takes a as a shape parameter for $a$ .

When  $a$ is an integer, gamma reduces to the Erlang distribution, and when $a = 1$ to the exponential distribution.

```python
# Choose a = 2
a = 2
# Сalculate the values of mean, variance, skew, and kurtosis of this distribution 
mu, var, skew, kurt = sts.gamma.stats(a, moments='mvsk')
```

```python
# Generate a sample of volume 1000 from the gamma distribution with parameter a = 2
r = sts.gamma.rvs(a, size=1000)
x = np.arange(0,10,0.05)

# Make a histogram of this sample and the theoretical probability density function
fig, ax = plt.subplots(1, 1,figsize=(10,6))
ax.plot(x, sts.gamma.pdf(x, a),'r-', lw=2, alpha=0.9, label='theoretical gamma pdf')
ax.hist(r, bins = 25, density=True, histtype='stepfilled', alpha=0.8, label = 'distribution of the sample')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
plt.legend()
ax.grid()
plt.show()
```

# Estimation of the distribution of a sample average random variable for different sample sizes

```python
# generate 1000 samples of volume n = 3 and count their sample averages
means_3 = []
for i in range(0,1000):
    sample = sts.gamma.rvs(a, size=3)
    means_3.append(sample.mean())
    
# calculate the values of the parameters of normal distributions, 
# which, according to the Central limit theorem, approximate the distribution of sample means, 
# using the theoretical mean and variance of our distribution 
sigma = np.sqrt(var/3)
normRv = sts.norm(loc=mu, scale=sigma)    
x = np.linspace(0,6,1000)
pdf = normRv.pdf(x)

# Make a histogram of sample means and a probability density function  
# of the normal distributions, which approximate it
fig, ax = plt.subplots(1, 1,figsize=(10,6))
ax.plot(x, pdf, label= 'normal pdf')
ax.hist(means_3, bins=15,density=True, histtype='stepfilled', alpha=0.8, label = 'distribution of the sample mean')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
ax.set_title('$n = 3$')
ax.grid()
plt.legend()
plt.show()
```

```python
# Let's do the same for n = 5

means_5 = []
for i in range(0,1000):
    sample = sts.gamma.rvs(a, size=5)
    means_5.append(sample.mean())

sigma = np.sqrt(var/5)
normRv = sts.norm(loc=mu, scale=sigma)    
x = np.linspace(0,5,1000)
pdf = normRv.pdf(x)

fig, ax = plt.subplots(1, 1,figsize=(10,6))
ax.plot(x, pdf, label= 'normal pdf')
ax.hist(means_5,bins=15, density=True, histtype='stepfilled', alpha=0.8, label = 'distribution of the sample mean')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
ax.set_title('$n = 5$')
ax.grid()
plt.legend()
plt.show()
```

```python
# n = 10
means_10 = []
for i in range(0,1000):
    sample = sts.gamma.rvs(a, size=10)
    means_10.append(sample.mean())

sigma = np.sqrt(var/10)
normRv = sts.norm(loc=mu, scale=sigma)    
x = np.linspace(0,5,1000)
pdf = normRv.pdf(x)

fig, ax = plt.subplots(1, 1,figsize=(10,6))
ax.plot(x, pdf, label= 'normal pdf')
ax.hist(means_10,bins=15, density=True, histtype='stepfilled', alpha=0.8, label = 'distribution of the sample mean')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
ax.set_title('$n = 10$')
ax.grid()
plt.legend()
plt.show()
```

```python
# n = 100
means_100 = []
for i in range(0,1000):
    sample = sts.gamma.rvs(a, size=100)
    means_100.append(sample.mean())

sigma = np.sqrt(var/100)
normRv = sts.norm(loc=mu, scale=sigma)    
x = np.linspace(1,3,1000)
pdf = normRv.pdf(x)

fig, ax = plt.subplots(1, 1,figsize=(10,6))

ax.plot(x, pdf,alpha=1, label= 'normal pdf')
ax.hist(means_100,bins=16, density=True, histtype='stepfilled', alpha=0.8, label = 'distribution of the sample mean')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
ax.set_title('$n = 100$')
ax.grid()
plt.legend()
plt.show()
```

# Conclusion

The distribution of the sample means is well approximated by a normal distribution with selected parameters, which confirms the Central limit theorem. Approximation accuracy increases with increasing of sample size.
