```python
may_2016['tpep_pickup_datetime'] = may_2016.tpep_pickup_datetime.apply(lambda x: x[:-3])
may_2016['tpep_dropoff_datetime'] = may_2016.tpep_dropoff_datetime.apply(lambda x: x[:-3])
```

```python
# Проведите фильтрацию: удалите поездки с
# нулевой длительностью
# нулевым количеством пассажиров
# нулевым расстоянием поездки по счётчику
# координатами начала, не попадающими в прямоугольник Нью-Йорка (от -74.25559 до -73.70001  и от 40.49612 до 40.91553)

may_2016 = may_2016[~(may_2016.passenger_count == 0)|
         (may_2016.tpep_pickup_datetime == may_2016.tpep_dropoff_datetime)|
         (may_2016.trip_distance == 0)|
         (may_2016.pickup_longitude < -74.25559)|
         (may_2016.pickup_longitude > -73.70001)|
         (may_2016.pickup_latitude < 40.49612)|
         (may_2016.pickup_latitude > 40.91553)]
```
